## Notes About Current Flood Mitigation and Resiliency Strategies

- Currently there are 7,400 miles of sewer pipes that collect sewage and stormwater
- 14 Wastewater Resource Recovery Facilities (WRRFs) that receive the flow
    - These systems help improve the cities health
- 60 percent of NYC's sewer system is combined, meaning that they combine both sanitary and storm flows
- When these sewers reach full capacity, CSO (combined sewer overflow) occours, causing thr water to appear in local waterways
- The last 40 percent sends sanitary sewage to WRRFs, and storm sewers send run off to waterbodies
- Improvements in sewage infrastructure and green infrastructure (45 billion dollar investment) decreased CSO by 80 percent since the 1960s 
- Another 10.6 billions dollars was invested to continue to reduce CSO
- Flood insurance for property owners
- Launching FloodHelpNY
    - Education program and webisite
    - Produces a newsletter notifying New Yorkers of flood risk and flood insurance and other updates
- The construction code
    - creates standards for flood resistant construction
        - Lowest floor with occupation must be above the designed flood elevation (DFE)
        - If a constructed area is below the DFE it can be used for parking, storage, etc.
        - Utilities must be above the DFE
- NYC Local Law of 2015
    - catch basins muct be cleaned every year instead of every 3 years
    - A malfunctioning catch basin must be fixed in nine days
- New Neighborhoods Special districts
    - limited the number of users in certain flood risk locations
- investing in sewer and drainage infrastructure, installing backflow valves, and increasing effectiveness of catch basin and storm drain maintenance. 
- Green Infrastructure 
    - Allows water to go underground easily and reduces the amount of stormwater and minimizes flooding
    - Green roofs help cool buildings and capture rain in barrels and reduce the amount of water going into sewers
    - rain gardens
- Protecting buildings
    - dry floodproofing: watertight constuction methods with temporary barriers
    - Wet floodproofing: Using flood resistant materials that let water in ot out without signifigant damage
    - Building elevation: Raising buildings above the DFE
    - Equipment elevation: raising mechanical equipment above the DFE
    - Land and street elevation
- Protecting infrastructure:
    - Subways: elevating air vents
    - tunnels: improving pumping capacity in subway, commuter train, and passenger car tunnels.
    - Subway and rail: Raising or dry floodproofing track switches and electrical equipment
    - Power plants:building powerplants near flood prone locations
    - Hospitals: Elevating electrical equipment and ensuring emergency backup power in hospitals.
